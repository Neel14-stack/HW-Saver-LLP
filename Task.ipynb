```python
l = []
from csv import DictReader
with open("df1.csv",'r',encoding="utf-8") as f:
    csv_dict_reader = DictReader(f)
    for row in csv_dict_reader:
        l.append(row['RAW DATA'])
        #print(row['RAW DATA'])
    
```

```python
a = []
for i in l:
    if i!='[]':
        a.append(i)
    
```

```python
pd.set_option('display.max_rows', None)
a
```

```python
az = []
for i in a:
    if len(i)>5:
        az.append(i)
```

```python
import re
q = []
for i in az:
    q.append(re.search(".\b", i))
```

```python
d=[]

for i in az:
    d.append(i[2:-2])

print(d)
```

```python
df = pd.DataFrame(d,columns = ['Technical Skills'])
```
